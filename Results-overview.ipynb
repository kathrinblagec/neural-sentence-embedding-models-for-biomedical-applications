## Correlations overview

```python
correlations_overview_df = cosines_overview_df.corr()
correlations_overview_df.to_csv(os.path.join(config.DATA_DIR, 'correlations-overview.csv'))
correlations_overview_df
```

```python
cosines_overview_df_converted = cosines_overview_df.iloc[:,4:]
# Convert ranges
for i in cosines_overview_df_converted.columns:
    range_value = 1
    old_min = np.min(cosines_overview_df_converted[str(i)])
    old_max = np.max(cosines_overview_df_converted[str(i)])
    old_range = old_max - old_min
    cosines_overview_df_converted[str(i)] = (cosines_overview_df_converted[str(i)] - old_min) / old_range
```

```python
list1 = [0, 1, 2]
list2 = [0, 1, 2]
list = [(x,y) for x in list1 for y in list2]

fig, (axes) = plt.subplots(3, 3, figsize=(12,12))
fig.text(0.4, 0.08, 'Estimated similarity score', size=16)
fig.text(0.06, 0.6, 'Annotated similarity score', rotation='vertical',  size=16)
for i, j in zip(cosines_overview_df_converted.columns.tolist(), list):
    sns.regplot(x=str(i), y=cosines_overview_df['Annotation scores'], data=cosines_overview_df_converted, ax=axes[j], color='#000000', scatter_kws={'s':1}, line_kws={"linewidth": 1, "color": "#000000"})
    axes[j].set_xlim(0.00, 1.00)
    axes[j].set_ylim(0.00, 4.00)
    axes[j].text(0.08, 0.87, 'r = ' + str(np.round(correlations_overview_df[i][1],3)), transform=axes[j].transAxes,bbox={'facecolor':'white', 'alpha':0.7, 'pad':6})
    axes[j].set_title(str(i))
    axes[j].set_xlabel('')
    axes[j].set_ylabel('')
```
